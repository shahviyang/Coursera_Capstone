In the cell below, I am using beautiful soup to webscrape postal codes data from a Wikipedia page, set as the variable url.

The code will print total number of tables and index of the table we are interested in.

```python
#retreiving the URL and creating a Beautiful soup object

url = 'https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'
data  = requests.get(url).text
soup = BeautifulSoup(data, 'html5lib')
tables = soup.find_all('table')
print('Total number of tables: '+str(len(tables)))

for index,table in enumerate(tables):
    if ("M1A" in str(table)):
        table_index = index
print('Index of table we want: '+str(table_index))
```

In the next cell, the table data in html format can be viewed.

```python
table = tables[table_index]
print(table.prettify())
```

In the next few cells, code that is already provided is used to make a dataframe out of the html format table printed in the cell before.

```python
'''
-> create a list
-> finding the table and table data
-> create a dictionary called cell having 3 keys PostalCode, Borough and Neighborhood
-> postal code contains upto 3 characters extract that using tablerow.p.text
-> use split ,strip and replace functions for getting Borough and Neighborhood information
-> Append to the list
'''
table_contents=[]
for row in table.findAll('td'):
    cell = {}
    if row.text[4:].split('\n')[0] == 'Not assigned':
        pass
    else:
        cell['PostalCode'] = row.text[:4].split('\n')[1]
        cell['Borough'] = (row.text[4:].split('\n')[0]).split('(')[0]
        cell['Neighborhood'] = (((((row.text[4:].split('\n')[0]).split('(')[1]).strip(')')).replace(' /',',')).replace(')',' ')).strip(' ')
        table_contents.append(cell)

        
#print(table_contents)
```

```python
#Create a dataframe with list
df=pd.DataFrame(table_contents)
df['Borough']=df['Borough'].replace({'Downtown TorontoStn A PO Boxes25 The Esplanade':'Downtown Toronto Stn A',
                                             'East TorontoBusiness reply mail Processing Centre969 Eastern':'East Toronto Business',
                                             'EtobicokeNorthwest':'Etobicoke Northwest','East YorkEast Toronto':'East York/East Toronto',
                                             'MississaugaCanada Post Gateway Processing Centre':'Mississauga'})
```

```python
df.to_csv('toronto-postal-data.csv')
df.head()
```
